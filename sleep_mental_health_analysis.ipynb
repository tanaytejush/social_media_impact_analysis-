```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr

sns.set(style="whitegrid")
%matplotlib inline
```

```python
df = pd.read_csv("Students Social Media Addiction.csv")
df.head()
```

```python
# Visualize Relationships
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x="Sleep_Hours_Per_Night", y="Addicted_Score", data=df, color="purple")
plt.title("Sleep Hours vs Addiction Score")
plt.xlabel("Sleep Hours Per Night")
plt.ylabel("Addicted Score")
plt.show()
```

```python
# Mental Health vs Addiction
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x="Mental_Health_Score", y="Addicted_Score", data=df, color="teal")
plt.title("Mental Health Score vs Addiction Score")
plt.xlabel("Mental Health Score")
plt.ylabel("Addicted Score")
plt.show()
```

```python
# Compute Correlation Coefficients , Pearson Corelation
```

```python
# Sleep vs Addiction
corr_sleep, p_sleep = pearsonr(df["Sleep_Hours_Per_Night"], df["Addicted_Score"])
print(f"Sleep vs Addiction: Correlation = {corr_sleep:.2f}, p-value = {p_sleep:.4f}")

# Mental Health vs Addiction
corr_mental, p_mental = pearsonr(df["Mental_Health_Score"], df["Addicted_Score"])
print(f"Mental Health vs Addiction: Correlation = {corr_mental:.2f}, p-value = {p_mental:.4f}")
```

```python
# Typical Expectations:
# Sleep vs Addiction: Likely negative correlation → More addiction = Less sleep
# Mental Health vs Addiction: Likely negative correlation → More addiction = Poorer mental health
```

```python
plt.figure(figsize=(8,6))
corr = df[["Sleep_Hours_Per_Night", "Mental_Health_Score", "Addicted_Score"]].corr()
sns.heatmap(corr, annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Heatmap: Sleep, Mental Health & Addiction")
plt.show()
```
