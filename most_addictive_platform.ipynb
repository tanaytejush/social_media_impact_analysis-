```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
%matplotlib inline
```

```python
df = pd.read_csv("Students Social Media Addiction.csv")
df.head()
```

```python
# Group Data by Platform
platform_addiction = df.groupby("Most_Used_Platform")["Addicted_Score"].mean().sort_values(ascending=False)
platform_addiction
```

```python
# Visualize the Results
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x=platform_addiction.index, y=platform_addiction.values, palette="coolwarm")
plt.title("Average Addiction Score by Platform", fontsize=16)
plt.ylabel("Average Addicted Score")
plt.xlabel("Most Used Platform")
plt.xticks(rotation=45)
plt.show()
```
